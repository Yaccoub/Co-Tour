```python
Englischer_Garten = pd.read_csv('Englischer Garten.csv',  header=0, parse_dates=["date"], squeeze=True)
```

```python
Englischer_Garten['date'] = Englischer_Garten['date'].replace({'Erlebnisdatum: ':''}, regex=True)
Englischer_Garten['visit'] = Englischer_Garten['visit'].replace({'Reiseart: ':''}, regex=True)
```

```python
Englischer_Garten[['city', 'country']] = Englischer_Garten['visitor_origin'].str.split(',', expand=True, n=2)
```

```python
Englischer_Garten.visit = Englischer_Garten.visit.fillna('allein')
```

```python
Englischer_Garten = pd.get_dummies(Englischer_Garten, columns=["visit", "country"])
```

```python
def preprocessing(dataframe):
    df = dataframe.copy()
    df['date'] = df['date'].replace({'Erlebnisdatum: ': ''}, regex=True)
    df['visit'] = df['visit'].replace({'Reiseart: ': ''}, regex=True)
    df[['city', 'country']] = df['visitor_origin'].str.split(',', expand=True, n=2)
    df = pd.get_dummies(df, columns=["visit", "country"])
    return df
```

```python
Englischer_Garten_post = preprocessing(Englischer_Garten)
```
