# Import dataset and Filter Target (Germany)

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv('/Users/yosrabahri/Documents/AMI/Mobility data/Global_Mobility_Report (1).csv',index_col=0)
```

```python
data.head()
```

```python
germany  = data[data.country_region == 'Germany'].copy()
germany.head()
```

# Pre-processing and Visualization

```python
germany.shape
```

```python
print(germany.isnull().sum())
```

We filter to see the non available data and clean it accordingly.
In this case, we delete 2 rows that are completely unavailable for Germany.

```python
germany.drop(["sub_region_2","census_fips_code"], axis =1)
```

Now, we start organising the data by month for a better visualization

```python
months_in_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
germany_mobility=germany.iloc[:,5:11];
germany_mobility
```

In the following graph, the fluctuations of the mobility in Germany are plotted from the 15th of February till the 19th of April. We divide the mobility according to the type of goal destination by color. We notice that the Parks percentage is the higher one.

```python
germany_mobility.plot(kind='line', x='date', figsize=(15,8), color=['blue', 'green', 'orange', 'purple', 'black'])
```
